```python
# Imports + settings 
import os
from datetime import datetime, timedelta

import numpy as np
import pandas as pd
from tqdm import tqdm

pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 140)

DATA_DIR_RAW = "../data/raw"
DATA_DIR_PROCESSED = "../data/processed"
os.makedirs(DATA_DIR_RAW, exist_ok=True)
os.makedirs(DATA_DIR_PROCESSED, exist_ok=True)
```

```python
# Download configuration
PAIR = "EURUSD"
START_DATE = "2024-01-01"   # adjust later
END_DATE   = "2024-03-01"   # start small for testing

RAW_OUT = os.path.join(DATA_DIR_RAW, f"{PAIR}_1min_{START_DATE}_{END_DATE}.csv")
PROCESSED_OUT = os.path.join(DATA_DIR_PROCESSED, f"{PAIR}_1min_{START_DATE}_{END_DATE}.parquet")

print("RAW_OUT:", RAW_OUT)
print("PROCESSED_OUT:", PROCESSED_OUT)
```

```python
# Download
import duka.app.app as duka_app
from duka.core.utils import TimeFrame
from duka.core.instruments import Instrument

# Parse dates
start_dt = datetime.fromisoformat(START_DATE)
end_dt = datetime.fromisoformat(END_DATE)

# Download to a folder (duka stores files; we’ll read into pandas after)
duka_download_dir = os.path.join(DATA_DIR_RAW, "duka_files")
os.makedirs(duka_download_dir, exist_ok=True)

# Run download
duka_app.download(
    instruments=[Instrument.EURUSD],
    start=start_dt,
    end=end_dt,
    timeframe=TimeFrame.M1,
    folder=duka_download_dir,
)

print("Download complete. Files in:", duka_download_dir)
```
